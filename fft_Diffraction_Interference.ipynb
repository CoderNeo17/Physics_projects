```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy.fft as fft

double_slit=np.zeros((400,500))
for x in range(0,500,1):
  for y in range(0,400,1):
    if ((x>=195 and x<=205) or (x>=295 and x<=305)) and (y>=50 and y<=350) :
      double_slit[y][x]=1
    else:
      double_slit[y][x]=0

plt.imshow(double_slit,cmap="gray",origin="lower")
```

```python
fourier_img=np.fft.fft2(double_slit)
final_img=(np.abs(np.fft.fftshift(fourier_img)))**2   #Intensity at a point is directly proportional to the square of amplitude of wave

plt.imshow(final_img,cmap="gray",norm="asinh",origin="lower") #asinh normalisation is used in every plot for visibility reasons
plt.title("Diffraction pattern of Double Slit setup")
```

```python
real_part=(np.real(np.fft.fftshift(fourier_img)))
plt.imshow(real_part,cmap="gray",norm="asinh",origin="lower")
plt.title("Plot of Real Part of the image")
```

```python
square_slit=np.zeros((400,500))
for x in range(0,500,1):
  for y in range(0,400,1):
    if x>=245 and x<=255 and y>=195 and y<=205 :
      square_slit[y][x]=1
    else:
      square_slit[y][x]=0

plt.imshow(square_slit,cmap="gray")
```

```python
fourier_img2=np.fft.fft2(square_slit)
final_img2=(np.abs(np.fft.fftshift(fourier_img2)))**2

plt.imshow(final_img2,cmap="gray",norm="asinh",origin="lower")
plt.title("Diffraction pattern of a Square Shaped slit")
```
