```python
import numpy as np
import matplotlib.pyplot as plt

data=np.genfromtxt('5000nearbystars.csv',delimiter=",")     #use your file name
data=data[1:,:]

def abs_g_band(arr):
  g_band=[]
  for i in range(5000):
    gband=arr[i][5]+5+5*np.log10(1/arr[i][4])
    g_band.append(gband)

  return g_band
```

```python
plt.figure(figsize=(7,9))
plt.scatter(data[:,1:2],abs_g_band(data),c=data[:,1:2],s=1,cmap="brg_r")
plt.axis([5,0,0,20])
plt.xlabel("BP-RP")
plt.ylabel("absolut g-magnitude")
plt.axis("auto")
plt.show()
```
